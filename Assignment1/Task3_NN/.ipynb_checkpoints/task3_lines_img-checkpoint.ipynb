```python
import tensorflow as tf
import matplotlib.pyplot as plt
import os
import numpy as np
import random
import cv2
```

```python
n_input = 2352   # input layer (28x28x3 pixels)
n_hidden1 = 800 # 1st hidden layer
# n_hidden2 = 256 # 2rd hidden layer
n_output = 96   # output layer (96 classes)
total_classes = 96 # 96 = 2 * 12 * 2 * 2
op_folder = "./LINE_IMG_DATA"
img_folder = "../Task1_Images/img"
img_in_each_class = 1000
training_samples = (img_in_each_class * 80) / 100
TOTAL_TESTING_SAMPLES = 9600
MAX_PIXEL_VALUE = 256.0
MODEL_SAVE_PATH = "./saved/model_lineImg"

learning_rate = 1e-2
n_iterations = 1000
batch_size = 96
```

```python
# Get class label, label => [0, total_classes - 1] 0 indexed
def getClassLabel(notation):
    l, m, n, k = (int(x) for x in notation.split("_"))
    label = l * (2 * 12 * 2) + m * (12 * 2) + n * (2) + k
    return int(label)
```

```python
# Reading the input classes
def getInputClasses():
    classes_folders = os.listdir(img_folder)
    arr = []
    for c in classes_folders:
        label = getClassLabel(c)
        arr.append([label, c])
    arr = np.array(arr)
    print arr
    return arr
```

```python
def getTrainingData(input_classes, batch_size):
    img_arr = []
    Y_arr = np.zeros([batch_size, total_classes])
    for i in range(batch_size):
        random_index = random.randint(0, total_classes - 1)
        label, class_folder = int(input_classes[random_index][0]), input_classes[random_index][1]
        img = cv2.imread(img_folder + "/" + class_folder + "/" + class_folder + "_" + `(random.randint(0, training_samples - 1))` + ".jpg")
        img = img.flatten()
        img = img / MAX_PIXEL_VALUE # Normalizing the pixels value to [0, 1]
        Y_arr[i][label] = 1 # label => [0, total_classes - 1] 0 indexed
        img_arr.append(img)
    img_arr = np.array(img_arr)
    return img_arr, Y_arr
#     return one_hot_notation, img
```

```python
def getTestingData(input_classes, batch_size):
    img_arr = []
    Y_arr = np.zeros([batch_size, total_classes])
    for i in range(batch_size):
        random_index = random.randint(0, total_classes - 1)
        label, class_folder = int(input_classes[random_index][0]), input_classes[random_index][1]
        img = cv2.imread(img_folder + "/" + class_folder + "/" + class_folder + "_" + `(random.randint(training_samples, img_in_each_class - 1))` + ".jpg")
        img = img.flatten()
        img = img / MAX_PIXEL_VALUE # Normalizing the pixels value to [0, 1]
        Y_arr[i][label] = 1 # label => [0, total_classes - 1] 0 indexed
        img_arr.append(img)
    img_arr = np.array(img_arr)
    return img_arr, Y_arr
```

```python
# Getting the input data and the ground truth in 1-hot notation
ip_data = tf.placeholder("float", [None, n_input])
Y = tf.placeholder("float", [None, n_output])

# Initializing the Weights matrices with a random normal distribution of S.D = 0.1
W = {
    'w1': tf.Variable(tf.truncated_normal([n_input, n_hidden1], stddev=0.1)),
    'w2': tf.Variable(tf.truncated_normal([n_hidden1, n_output], stddev=0.1))
}

# Initializing the biases
b = {
    'b1': tf.Variable(tf.constant(0.2, shape=[n_hidden1])),
    'b2': tf.Variable(tf.constant(0.2, shape=[n_output]))
}

a1 = tf.add(tf.matmul(ip_data, W['w1']), b['b1'])
a1 = tf.nn.relu(a1)
a2 = tf.add(tf.matmul(a1, W['w2']), b['b2'])
# a2 = tf.nn.softmax(a2)
# a3 = tf.add(tf.matmul(a2, W['w3']), b['b3'])
# a3 = tf.nn.softmax(a3)

# Cost function
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=Y, logits=a2))
train_epoch = tf.train.AdamOptimizer(learning_rate).minimize(cross_entropy)

correct_pred = tf.equal(tf.argmax(a2, 1), tf.argmax(Y, 1))
accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))

prediction = tf.argmax(a2, 1)
labels = tf.argmax(Y, 1)
conf_matrix = tf.math.confusion_matrix(labels, prediction)
precision = tf.metrics.precision(labels, prediction)
recall = tf.metrics.recall(labels, prediction)
```

```python
# Training Phase

Loss, Accuracy, itr = [], [], []
init_op = tf.group(tf.global_variables_initializer(), tf.local_variables_initializer())

model_saver = tf.train.Saver()

sess = tf.Session()
sess.run(init_op)

input_classes = getInputClasses()

# training the model
for i in range(n_iterations):
    itr.append(i + 1)
    print i+1,
    batch_X, batch_Y = getTrainingData(input_classes, batch_size)
    sess.run(train_epoch, feed_dict={ip_data: batch_X, Y: batch_Y})
    mini_loss, mini_accuracy = sess.run([cross_entropy, accuracy], feed_dict={ip_data: batch_X, Y: batch_Y})
    Loss.append(mini_loss)
    Accuracy.append(mini_accuracy)
    if(i % 10 == 0):
        print "\n" + str(i + 1) + " Loss: " + str(mini_loss) + " Accuracy: " + str(mini_accuracy * 100.0)
save_path = model_saver.save(sess, MODEL_SAVE_PATH)
print("Model saved in path: %s" % save_path)
plt.plot(itr, Accuracy)
plt.xlabel("Iteration")
plt.ylabel("Accuracy")
plt.savefig(op_folder + "/iteration_accuracy.png", bbox_inches='tight')
plt.show()
# sess.close()
```
